```julia
# Bibliotecas usadas
using DifferentialEquations, Plots
```

```julia
# Parâmetros
sigma = 1.0
x0 = 0.0
tspan = (0.0, 20.0)

# Drift e difusão
f1(X, p, t) = X - X^3
f2(X, p, t) = sigma

W = WienerProcess(0.0, 0.0, 0.0)

prob = SDEProblem(f1, f2, x0, tspan, noise=W)

sol = solve(prob, SRIW1(), dt=0.01)

plot(sol.t, sol.u;
     label = "X(t)",
     xlabel = "Tempo",
     ylabel = "X",
     title = "Modelo estocástico",
     size = (960, 540),
     color = :red,
     linewidth = 1.5)
```
